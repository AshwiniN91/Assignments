### Transformation of iris data into 3 dimensions and plot a 3d chart with transformed dimensions and colour each data point with specific class.

```python
# Importing libraries
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
def scatter_plot(x_label,y_label,z_label,tgt,c,m,label):
    x = iris[ iris['species'] == tgt ][x_label]
    y = iris[ iris['species'] == tgt ][y_label]
    z = iris[ iris['species'] == tgt ][z_label]
    fig.scatter(x,y,z,color=c,edgecolors='k',s=50,alpha=0.9,marker=m,label=label)
    fig.set_xlabel(x_label)
    fig.set_ylabel(y_label)
    fig.set_zlabel(z_label)
    return

fig = plt.figure(figsize=(10,8)).gca(projection='3d') 
scatter_plot('sepal_length','sepal_width','petal_length','setosa','g','o','setosa')
scatter_plot('sepal_length','sepal_width','petal_length','versicolor','b','o','versicolor')
scatter_plot('sepal_length','sepal_width','petal_length','virginica','r','o','virginica')
plt.title("3-D chart of Iris dataset",fontsize=15)
plt.legend()
plt.show()
```
